```python
import pandas as pd

# Load the dataset
df = pd.read_csv('./dataset.csv')

# Fill missing values in 'Sr. No' column
df['Sr. No'] = df['Sr. No'].fillna(method='ffill')

# Group by 'Sr. No'
grouped_df = df.groupby('Sr. No')

# Required columns
required_columns = ['Data'] + [f'Unnamed: {i}' for i in range(3, 10)]

# Process each group
processed_groups = {}
for name, group in grouped_df:
    missing_columns = [
        col for col in required_columns if col not in group.columns]
    if missing_columns:
        print(f"Missing columns in group {name}: {missing_columns}")
        continue
    filtered_group = group[required_columns].copy()
    filtered_group.fillna('', inplace=True)
    processed_groups[name] = filtered_group

# Drop required columns from the original dataframe
df.drop(columns=required_columns, inplace=True)

# Create new dataframe with first entries from each group
new_df = df.groupby('Sr. No').first()

# Process each group to handle varying lengths of columns
for i in range(1, len(processed_groups) + 1):
    d = processed_groups[i].copy()
    new_columns = d.iloc[0]
    d.columns = new_columns
    d = d.drop(d.index[0])
    d.reset_index(drop=True, inplace=True)
    d.replace("", float("NaN"), inplace=True)

    # Remove all NaN columns
    d.dropna(how='all', axis=1, inplace=True)

    # Ensure to drop only fully NaN rows if they exist
    d.dropna(how='all', inplace=True)

    # Align columns by filling NaNs at the end
    d = d.apply(lambda x: pd.Series(x.dropna().values))

    # Assign the processed group back
    processed_groups[i] = d

# Assign the processed groups' values to the 'Data' column
new_df = new_df.assign(Data=list(processed_groups.values()))

# Reset index and drop 'Sr. No' column
df = new_df.reset_index().drop(columns='Sr. No')

df.head()
```

```python
df['rows'] = df['Data'].apply(lambda x: x.shape[0])
df['columns'] = df['Data'].apply(lambda x: x.shape[1])
df.head(10)
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')

for i in range(len(df)):
    headers = df.iloc[i]['Data'].columns
    headers = [
        header.split()[0] if ' ' in header else header for header in headers]
    numeric_headers = [int(header) for header in headers if header.isdigit()]
    max_header = max(numeric_headers) if numeric_headers else None
    headers = [header if not header.isdigit() else str(max_header)
               for header in headers]
    header_embeddings = model.encode(headers)
    similarity_matrix = cosine_similarity(header_embeddings)
    np.fill_diagonal(similarity_matrix, 0)
    max_similarity = np.max(similarity_matrix, axis=1)
    distinct_entities = np.sum(max_similarity < 0.7)
    df.at[i, 'distinct_entities_count'] = distinct_entities + 1
```

```python
df['isNested'] = False

for index, row in df.iterrows():
    sample = row['Data'].dropna(how='any')
    
    nested = True
    for i in range(len(sample)):
        if sample.iloc[i].iloc[0] != sample.iloc[0].iloc[0]:
            nested = False
            break

    if nested:
        df.at[index, 'isNested'] = True
```
